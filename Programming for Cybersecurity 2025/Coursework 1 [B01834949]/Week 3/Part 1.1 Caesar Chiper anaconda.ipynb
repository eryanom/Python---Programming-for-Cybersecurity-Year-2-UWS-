```python
# Erya Anom
# Lab - Caesar Cipher Encryption
# This program encrypts text using the Caesar cipher method
# by shifting letters in the alphabet by a user-specified amount.

import sys  # Import sys for clean program exit

# Define the alphabet used for encryption
alphabet = 'abcdefghijklmnopqrstuvwxyz'
# Function: encrypt()
def encrypt(plaintext, k):
    """
    Encrypt text using Caesar cipher.
    plaintext: text to encrypt
    k: shift value (number of letters to move in the alphabet)
    """
    ciphertext = ''  # Empty string to store encrypted text
    print("\n--- Encryption Process ---")
    # Loop through every character in the plaintext (converted to lowercase)
    for char in plaintext.lower():
        if char in alphabet:
            old_index = alphabet.index(char)
            new_index = (old_index + k) % 26  # Wrap around alphabet with modulo
            new_char = alphabet[new_index]

            # Show encryption steps
            print(f"'{char}' -> index {old_index} -> shifted {k} -> new index {new_index} -> '{new_char}'")
            ciphertext += new_char
        else:
            # Keep non-letter characters unchanged
            print(f"'{char}' (non-letter) remains unchanged")
            ciphertext += char

    return ciphertext


# Main program loop
while True:
    # Ask the user for input text
    plaintext = input("\nEnter your plaintext (or type 'stop' to quit): ").strip()

    # Check if user wants to exit
    if plaintext.lower() == "stop":
        confirm = input("Are you sure you want to quit? (yes/no): ").strip().lower()
        if confirm in ("yes", "y"):
            print("Goodbye! ðŸ‘‹ Program exited successfully.")
            sys.exit()  # Cleanly terminate the program
        else:
            print("Continuing program...\n")
            continue  # Go back to asking for input

    # Ask for shift value
    try:
        shift = int(input("Enter shift value (e.g., 1, 2, 3...): "))
    except ValueError:
        print("âš ï¸ Please enter a valid number for the shift.")
        continue

    # Encrypt the plaintext
    ciphertext = encrypt(plaintext, shift)

    # Display final results
    print("\nPlaintext :", plaintext)
    print("Ciphertext:", ciphertext)
    print("-" * 40)
```
